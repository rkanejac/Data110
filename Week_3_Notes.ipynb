```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importing data
df = pd.read_csv("https://raw.githubusercontent.com/Reben80/Data110-32213/main/dataset/tempnormals.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#shows temp for multiple locations
plt.plot(df['day_of_year'],df['temperature'])
```

```python
#trying to show data for one location = death valley
dv = df[df['location']=='Death Valley']
plt.plot(dv['day_of_year'],dv['temperature'],label = 'Death Valley')
plt.legend()
plt.grid()
plt.xlabel('Day of the Year')
plt.ylabel('Temperature')
plt.show()
```

```python
DV_stat = dv['temperature'].describe()
Avg_dv = DV_stat['mean']
print(Avg_dv)
```

```python
dv = df[df['location']=='Death Valley']
plt.plot(dv['day_of_year'],dv['temperature'],label = 'Death Valley')
plt.axhline(y=Avg_dv, color='red')
plt.legend()
plt.grid()
plt.xlabel('Day of the Year')
plt.ylabel('Temperature')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,3))

location_order_reversed = ['Chicago', 'San Diego', 'Houston', 'Death Valley']

heatmap_data = df.groupby(['location', 'month'])['temperature'].mean().unstack(level=1)
heatmap_data_reversed = heatmap_data.reindex(location_order_reversed)

sns.heatmap(heatmap_data_reversed, cmap="YlOrRd", linewidths=0.5, annot=False)
plt.title('Average Monthly Temperature Normals by Location')
plt.xlabel('Month')
plt.ylabel('Location')
plt.show()
```

```python
#simple heat map
test_data = np.array([[1,2,3],[3,4,5]])
sns.heatmap(test_data, linewidths=0.5)
```

```python

plt.figure(figsize=(14,6))

#groups by the location
grouped = df.groupby('location')
print(grouped)

for name, group in grouped:
  plt.plot(group['day_of_year'], group['temperature'], label=name)

plt.title('Temperature Trends by Location Throughout the Year')
plt.xlabel('Day of the Year')
plt.ylabel('Temperature')
plt.legend()
plt.show()

```

```python
#graphing bar charts
x = [1, 2, 3, 4]
y = [10, 8, 12, 16]
plt.bar(x, y, label='Week 1')
plt.legend()
plt.show()
```

```python
#If you want to compare two bar charts you have to compare them side by side
x = ['p1', 'p2', 'p3', 'p4']
y = [10, 8, 12, 16]
z = [12, 7, 8, 23]
s = [2, 22, 7, 6]
plt.bar(x, y, label='Week 1')
plt.bar(x, z, label='Week 2', bottom=y)  # Stacks z on top of y
plt.title('Wrong Bar Chart')
plt.legend()
plt.show()
```

```python
x_indexes = np.arange(len(x)) #number of indices
plt.bar(x_indexes - 0.125, y, width=0.25, label='Week 1')
plt.bar(x_indexes + 0.125, z, width=0.25, label='Week 2')
plt.bar(x_indexes + 0.37, s, width=0.25, label='Week 3')
plt.xticks(ticks=x_indexes, labels=x)
plt.title('Right Bar Chart')
plt.legend()
plt.show()
```
