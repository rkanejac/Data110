```python
#importing data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


df = pd.read_csv("/content/us_gun_deaths_2000s.csv")
guns = pd.DataFrame(df)

gun_deaths_per_year = pd.DataFrame(columns = ['Year', 'Count'], index=range(19))
gun_deaths_per_year['Year'] = guns['year'].unique()
gun_deaths_per_year['Year'] = gun_deaths_per_year['Year']. astype('str')
years = guns['year'].unique()

year_counts = {}
# Loop through each year in the DataFrame
for year in years:
  count = guns[guns['year'] == year].shape[0]
  year_counts[year] = count

gun_deaths_per_year['Count'] = year_counts.values()
print('\n')
#print(gun_deaths_per_year)


# Create a bar plot
plt.figure(figsize=(15,8))
plot = plt.bar(gun_deaths_per_year['Year'].astype(str), gun_deaths_per_year['Count'], color='mistyrose', edgecolor='maroon')
plt.xlabel('Year')
plt.ylabel('Number of Gun Deaths')
plt.title('Gun Deaths in the US from 2000 to 2018', fontsize=20)
ax = plt.gca()
ax.set_facecolor("white")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.bar_label(plot, labels = year_counts.values(), label_type = 'edge')
plt.show()
```

```python
#race analysis
gun_deaths_per_race = pd.DataFrame(columns = ['Race', 'Count'], index=range(6))
gun_deaths_per_race['Race'] = guns['victim_race_plus_hispanic'].unique()
gun_deaths_per_race['Race'] = gun_deaths_per_race['Race'].astype('str')
races = guns['victim_race_plus_hispanic'].unique()

races_count = {}
for race in races:
  count = guns[guns['victim_race_plus_hispanic'] == race].shape[0]
  races_count[race] = count

gun_deaths_per_race['Count'] = races_count.values()
gun_deaths_per_race = gun_deaths_per_race.sort_values('Count')
#print(gun_deaths_per_race)

plt.figure(figsize=(15,8))
plt.barh(gun_deaths_per_race['Race'].astype(str), gun_deaths_per_race['Count'], color='mistyrose', edgecolor='maroon')
plt.ylabel('Race', fontsize=15)
plt.xlabel('Number of Gun Deaths', fontsize=15)
plt.title('Gun Deaths in the US by Race', fontsize=20)
ax = plt.gca()
ax.set_facecolor("white")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_axisbelow(True)
plt.grid(axis='x')
plt.show()

```

```python
#gender analysis
gun_deaths_per_gender = pd.DataFrame(columns = ['Gender', 'Count'], index=range(3))
gun_deaths_per_gender['Gender'] = guns['victim_sex'].unique()
gun_deaths_per_gender['Gender'] = gun_deaths_per_gender['Gender'].astype('str')
genders = guns['victim_sex'].unique()

gender_count = {}
for gender in genders:
  count = guns[guns['victim_sex'] == gender].shape[0]
  gender_count[gender] = count

gun_deaths_per_gender['Count'] = gender_count.values()
gun_deaths_per_gender = gun_deaths_per_gender.sort_values('Count', ascending=False)
print(gun_deaths_per_gender)

plt.figure(figsize=(15,8))
plot2 = plt.bar(gun_deaths_per_gender['Gender'].astype(str), gun_deaths_per_gender['Count'], color='mistyrose', edgecolor='maroon')
plt.xlabel('Gender', fontsize=15)
plt.ylabel('Number of Gun Deaths', fontsize =15)
plt.title('Gun Deaths in the US by Gender')



ax = plt.gca()
ax.set_facecolor("white")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_axisbelow(True)
plt.grid(axis='y')
plt.bar_label(plot2, labels = gender_count.values(), label_type = 'edge')
plt.show()
```

```python
#age distribution analysis
guns['victim_age'] = guns['victim_age'].fillna(value=-10)

#replaced victims younger than 1 with 0.5 and victims older than 99 with 100
guns['victim_age'] = guns['victim_age'].replace('Less than one year old', -1)
guns['victim_age'] = guns['victim_age'].replace('99 years or older', 100)
guns['victim_age'] = guns['victim_age'].astype('i')
plt.hist(guns['victim_age'], bins = 15, edgecolor ='white')
plt.xticks(np.arange(0, 110, step=10))
plt.show()
```
