```python
## your code for task 1 should be here.
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
x = np.arange(-2*np.pi, 2 * np.pi, np.pi/24) #ask about graph not touching
plt.xticks(ticks = [-2*np.pi,-np.pi,0,np.pi,2*np.pi], labels = ['-2π', '-π','0','π','2π'])
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, label = 'Sine Wave', color = 'red')
plt.plot(x,y2, label = 'Cosine Wave', color = 'purple')
plt.legend(loc='lower left')
plt.title('Cosine and Sine Waves')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
# Your code for Task 2 should be here

import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers from a normal distribution
data = np.random.normal(0, 1, 1000)
avg = np.average(data)
print('The average is:')
print(avg)

# Plot histogram
plt.hist(data,bins=40, color='gold')
plt.axvline(x=avg, color='green', label='Mean Line')
plt.legend()
plt.plot(avg)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

```

```python
## Your code for Task 3 should be here

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



# File path
file_path = 'happiness_2017.csv'

# Reading the CSV file
dat = pd.read_csv(file_path)
data = pd.DataFrame(dat)

# Displaying the first few rows of the dataframe
print('\n Imported happiness_2017 file using Local CVS File')
print(data.head())

#displaying the first few rows
dv = pd.read_csv("https://raw.githubusercontent.com/Reben80/Data110-32213/main/dataset/tempnormals.csv")
df = pd.DataFrame(dv)
print('\n')
print('\n Imported tempnormals file using a Link')
print(df.head())

#Combined Data Analysis

# Numerical Data Part
happymean = np.mean(data['HappinessScore'])
happymedian = np.median(data['HappinessScore'])
happystdv = np.std(data['HappinessScore'])
print('\n')
print('Data Analysis of the Happiness_2017 File')
print('1.) The mean of the HappinessScore Column is:',happymean)
print('2.) The median of the HappinessScore Column is:',happymedian)
print('3.) The standard deviation of the HappinessScore Column is:', happystdv)


tempmean = np.mean(df['temperature'])
tempmedian = np.median(df['temperature'])
tempstdv = np.std(df['temperature'])
print('\n')
print('Data Analysis of the Tempnormal File')
print('1.) The mean of the Temperature Column is:',tempmean)
print('2.) The median of the Temperature Column is:',tempmedian)
print('3.) The standard deviation of the Temperature Column is:', tempstdv)

# Categorical Data Part

# print('\n')
# print('Value Counts:')
# grouped1 = data['Region'].value_counts()
# print(grouped1)

print('\n')
print('Categorical Frequency of Regions in happiness_2017 File')
grouped = data.groupby('Region').size()
print(grouped)

print('\n')
print('Categorical Frequency of Locations in Tempnormals file')
grouped1 = df.groupby('location').size()
print(grouped1)

```
